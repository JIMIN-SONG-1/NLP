## 요약의 종류

1. Abstractive(Human-like) summarization

> Abstractive summarization(생성 요약)은 기존 Input text를 그대로 인용하지 않고, 기존의 내용을 새롭게 re-phrasing 하여 Summary를 생성하는 요약 모델입니다.

2. Extractive summarization

> 반면에 Extractive summarization(추출 요약)은 기존 Input text에 존재하는 중요한 단어를 그대로 사용하여 Summary를 생성하는 요약 모델입니다.

````python
from googletrans import Translator
import json

def translate_text(text, target_language='ko'):
    translator = Translator()
    # text가 리스트가 아닌 문자열인 경우를 처리
    if isinstance(text, str):
        # 문장 단위로 나누기 (마침표 기준)
        sentences = text.split('. ')
        translated_sentences = []
        for sentence in sentences:
            if sentence.strip():  # 빈 문장 제외
                translated_sentence = translator.translate(sentence, dest=target_language)
                translated_sentences.append(translated_sentence.text)
        return '. '.join(translated_sentences)
    else:
        # 이미 리스트인 경우
        translated_sentences = []
        for sentence in text:
            translated_sentence = translator.translate(sentence, dest=target_language)
            translated_sentences.append(translated_sentence.text)
        return translated_sentences

# 응답받은 텍스트 디코딩
response_text = response.text
response_json = json.loads(response_text)

# 번역할 텍스트 추출 (TLDRThis API는 보통 'summary' 키를 사용)
text_to_translate = response_json.get('summary', '')

# 번역된 텍스트
translated_text = translate_text(text_to_translate)

print("원본 요약:", text_to_translate)
print("번역된 텍스트:", translated_text)
````

````python
import requests
import time
from googletrans import Translator
from pprint import pprint

# API 키 설정
rapidapi_key = '2e2b83758cmshc941452c612bd1ap1683a7jsn438e555262ab'

def summarize_paper(article_url, min_length=100, max_length=300, is_detailed=False):
    """
    TLDRThis API를 사용하여 논문/기사를 요약하는 함수

    Args:
        article_url (str): 요약할 논문/기사의 URL
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        is_detailed (bool): 상세한 요약 여부

    Returns:
        dict: 요약 결과 또는 에러 메시지
    """
    url = "https://tldrthis.p.rapidapi.com/v1/model/extractive/summarize-url/"

    payload = {
        "url": article_url,
        "min_length": min_length,
        "max_length": max_length,
        "is_detailed": is_detailed
    }

    headers = {
        "content-type": "application/json",
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "tldrthis.p.rapidapi.com"
    }

    try:
        response = requests.post(url, json=payload, headers=headers)
        response_data = response.json()

        print(f"API 응답 상태: {response.status_code}")
        print(f"API 응답: {response_data}")

        # 에러 처리
        if 'detail' in response_data:
            return {"success": False, "error": response_data['detail']}

        if 'message' in response_data:
            return {"success": False, "error": response_data['message']}

        # 요약 데이터 추출
        if 'summary' in response_data:
            summary_data = response_data['summary']
        elif 'result' in response_data:
            summary_data = response_data['result']
        elif 'data' in response_data:
            summary_data = response_data['data']
        else:
            return {
                "success": False,
                "error": "요약 데이터를 찾을 수 없습니다",
                "available_keys": list(response_data.keys())
            }

        # 요약 텍스트 정리
        if isinstance(summary_data, list):
            # 리스트의 모든 요소를 합치기
            summary = " ".join([str(item).strip() for item in summary_data if item]).strip()
        else:
            summary = str(summary_data).strip()

        return {"success": True, "summary": summary}

    except requests.exceptions.RequestException as e:
        return {"success": False, "error": f"네트워크 오류: {str(e)}"}
    except Exception as e:
        return {"success": False, "error": f"예상치 못한 오류: {str(e)}"}

def translate_text(text, target_language='ko', source_language='en'):
    """
    구글 번역을 사용하여 텍스트를 번역하는 함수

    Args:
        text (str): 번역할 텍스트
        target_language (str): 대상 언어 (기본값: 한국어)
        source_language (str): 원본 언어 (기본값: 영어)

    Returns:
        dict: 번역 결과 또는 에러 메시지
    """
    try:
        translator = Translator()
        translated = translator.translate(text, src=source_language, dest=target_language)

        return {
            "success": True,
            "original_text": text,
            "translated_text": translated.text,
            "detected_language": translated.src
        }

    except Exception as e:
        return {"success": False, "error": f"번역 오류: {str(e)}"}

def summarize_and_translate(article_url, min_length=100, max_length=300, is_detailed=False):
    """
    논문/기사를 요약하고 한국어로 번역하는 통합 함수

    Args:
        article_url (str): 요약할 논문/기사의 URL
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        is_detailed (bool): 상세한 요약 여부

    Returns:
        dict: 요약 및 번역 결과
    """
    print(f"처리 중인 URL: {article_url}")
    print("=" * 50)

    # 1단계: 요약
    summary_result = summarize_paper(article_url, min_length, max_length, is_detailed)

    if not summary_result["success"]:
        return {
            "success": False,
            "url": article_url,
            "error": summary_result["error"]
        }

    # 2단계: 번역
    translation_result = translate_text(summary_result["summary"])

    if not translation_result["success"]:
        return {
            "success": False,
            "url": article_url,
            "summary": summary_result["summary"],
            "translation_error": translation_result["error"]
        }

    return {
        "success": True,
        "url": article_url,
        "original_summary": translation_result["original_text"],
        "translated_summary": translation_result["translated_text"],
        "detected_language": translation_result["detected_language"]
    }

def batch_summarize_and_translate(urls, min_length=100, max_length=300, delay=2):
    """
    여러 URL을 일괄 처리하는 함수 (Rate limit 방지를 위한 지연 포함)

    Args:
        urls (list): URL 리스트
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        delay (int): URL 간 처리 지연 시간 (초)

    Returns:
        list: 각 URL의 처리 결과 리스트
    """
    results = []

    for i, url in enumerate(urls):
        print(f"\n[{i+1}/{len(urls)}] 처리 중...")
        result = summarize_and_translate(url, min_length, max_length)
        results.append(result)

        # 마지막 URL이 아니면 지연
        if i < len(urls) - 1:
            print(f"{delay}초 대기 중...")
            time.sleep(delay)

    return results

# 사용 예시
if __name__ == "__main__":
    # 단일 URL 테스트
    test_url = "https://arxiv.org/html/2505.24090v1"
    result = summarize_and_translate(test_url, 50, 150)
    pprint(result)

    # 여러 URL 테스트 (Rate limit 주의)
    # test_urls = [
    #     "https://arxiv.org/html/2505.24090v1",
    #     "https://proceedings.mlr.press/v97/tan19a.html"
    # ]
    # results = batch_summarize_and_translate(test_urls, delay=5)
    # for result in results:
    #     pprint(result)
````

````python
import requests
import time
from googletrans import Translator
from pprint import pprint

# API 키 설정
rapidapi_key = '2e2b83758cmshc941452c612bd1ap1683a7jsn438e555262ab'

def summarize_paper(article_url, min_length=100, max_length=300, is_detailed=False, method="extractive"):
    """
    TLDRThis API를 사용하여 논문/기사를 요약하는 함수

    Args:
        article_url (str): 요약할 논문/기사의 URL
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        is_detailed (bool): 상세한 요약 여부
        method (str): "extractive" 또는 "abstractive"

    Returns:
        dict: 요약 결과 또는 에러 메시지
    """
    # 요약 방식에 따라 URL 설정
    if method == "abstractive":
        url = "https://tldrthis.p.rapidapi.com/v1/model/abstractive/summarize-url/"
    else:
        url = "https://tldrthis.p.rapidapi.com/v1/model/extractive/summarize-url/"

    payload = {
        "url": article_url,
        "min_length": min_length,
        "max_length": max_length,
        "is_detailed": is_detailed
    }

    headers = {
        "content-type": "application/json",
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "tldrthis.p.rapidapi.com"
    }

    try:
        response = requests.post(url, json=payload, headers=headers)
        response_data = response.json()

        print(f"API 응답 상태: {response.status_code}")
        print(f"요약 방식: {method.upper()}")
        print(f"API 응답: {response_data}")

        # 에러 처리
        if 'detail' in response_data:
            return {"success": False, "error": response_data['detail']}

        if 'message' in response_data:
            return {"success": False, "error": response_data['message']}

        # 요약 데이터 추출
        if 'summary' in response_data:
            summary_data = response_data['summary']
        elif 'result' in response_data:
            summary_data = response_data['result']
        elif 'data' in response_data:
            summary_data = response_data['data']
        else:
            return {
                "success": False,
                "error": "요약 데이터를 찾을 수 없습니다",
                "available_keys": list(response_data.keys())
            }

        # 요약 텍스트 정리
        if isinstance(summary_data, list):
            # 리스트의 모든 요소를 합치기
            summary = " ".join([str(item).strip() for item in summary_data if item]).strip()
        else:
            summary = str(summary_data).strip()

        return {"success": True, "summary": summary, "method": method}

    except requests.exceptions.RequestException as e:
        return {"success": False, "error": f"네트워크 오류: {str(e)}"}
    except Exception as e:
        return {"success": False, "error": f"예상치 못한 오류: {str(e)}"}

def translate_text(text, target_language='ko', source_language='en'):
    """
    구글 번역을 사용하여 텍스트를 번역하는 함수

    Args:
        text (str): 번역할 텍스트
        target_language (str): 대상 언어 (기본값: 한국어)
        source_language (str): 원본 언어 (기본값: 영어)

    Returns:
        dict: 번역 결과 또는 에러 메시지
    """
    try:
        translator = Translator()
        translated = translator.translate(text, src=source_language, dest=target_language)

        return {
            "success": True,
            "original_text": text,
            "translated_text": translated.text,
            "detected_language": translated.src
        }

    except Exception as e:
        return {"success": False, "error": f"번역 오류: {str(e)}"}

def summarize_and_translate(article_url, min_length=100, max_length=300, is_detailed=False, method="extractive"):
    """
    논문/기사를 요약하고 한국어로 번역하는 통합 함수

    Args:
        article_url (str): 요약할 논문/기사의 URL
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        is_detailed (bool): 상세한 요약 여부
        method (str): "extractive" 또는 "abstractive"

    Returns:
        dict: 요약 및 번역 결과
    """
    print(f"처리 중인 URL: {article_url}")
    print("=" * 50)

    # 1단계: 요약
    summary_result = summarize_paper(article_url, min_length, max_length, is_detailed, method)

    if not summary_result["success"]:
        return {
            "success": False,
            "url": article_url,
            "method": method,
            "error": summary_result["error"]
        }

    # 2단계: 번역
    translation_result = translate_text(summary_result["summary"])

    if not translation_result["success"]:
        return {
            "success": False,
            "url": article_url,
            "method": method,
            "summary": summary_result["summary"],
            "translation_error": translation_result["error"]
        }

    return {
        "success": True,
        "url": article_url,
        "method": method,
        "original_summary": translation_result["original_text"],
        "translated_summary": translation_result["translated_text"],
        "detected_language": translation_result["detected_language"]
    }

def compare_methods(article_url, extractive_max=500, abstractive_max=300, delay=5):
    """
    Extractive와 Abstractive 방식을 비교하는 함수

    Args:
        article_url (str): 요약할 논문/기사의 URL
        extractive_max (int): Extractive 방식의 최대 길이 (500까지 가능)
        abstractive_max (int): Abstractive 방식의 최대 길이 (300까지만 가능)
        delay (int): 두 요청 사이의 지연 시간 (초)

    Returns:
        dict: 두 방식의 비교 결과
    """
    print("=" * 60)
    print("EXTRACTIVE vs ABSTRACTIVE 비교")
    print("=" * 60)

    # Extractive 방식
    print(f"\n[1] EXTRACTIVE 방식 (원문 문장 추출, 최대 {extractive_max}자):")
    extractive_result = summarize_and_translate(article_url, 200, extractive_max, method="extractive")

    print(f"\n{delay}초 대기 중... (Rate limit 방지)")
    time.sleep(delay)

    # Abstractive 방식
    print(f"\n[2] ABSTRACTIVE 방식 (AI 재작성, 최대 {abstractive_max}자):")
    abstractive_result = summarize_and_translate(article_url, 100, abstractive_max, method="abstractive")

    return {
        "url": article_url,
        "extractive": extractive_result,
        "abstractive": abstractive_result
    }

def batch_summarize_and_translate(urls, min_length=100, max_length=300, method="extractive", delay=2):
    """
    여러 URL을 일괄 처리하는 함수 (Rate limit 방지를 위한 지연 포함)

    Args:
        urls (list): URL 리스트
        min_length (int): 요약의 최소 길이
        max_length (int): 요약의 최대 길이
        method (str): "extractive" 또는 "abstractive"
        delay (int): URL 간 처리 지연 시간 (초)

    Returns:
        list: 각 URL의 처리 결과 리스트
    """
    results = []

    for i, url in enumerate(urls):
        print(f"\n[{i+1}/{len(urls)}] 처리 중...")
        result = summarize_and_translate(url, min_length, max_length, method=method)
        results.append(result)

        # 마지막 URL이 아니면 지연
        if i < len(urls) - 1:
            print(f"{delay}초 대기 중...")
            time.sleep(delay)

    return results

# 사용 예시
if __name__ == "__main__":
    # 테스트 URL
    test_url = "https://arxiv.org/html/2505.24090v1"

    # 방법 1: Extractive 방식 (기본)
    print("\n=== EXTRACTIVE 방식 테스트 ===")
    result_extractive = summarize_and_translate(test_url, 200, 500, method="extractive")
    pprint(result_extractive)

    # 잠시 대기
    print("\n5초 대기 중...")
    time.sleep(5)

    # 방법 2: Abstractive 방식 (최대 300자 제한)
    print("\n=== ABSTRACTIVE 방식 테스트 (최대 300자) ===")
    result_abstractive = summarize_and_translate(test_url, 100, 300, method="abstractive")
    pprint(result_abstractive)

    # 방법 3: 두 방식 비교 (시간이 더 걸림)
    # comparison = compare_methods(test_url, extractive_max=500, abstractive_max=300, delay=10)
    # print("\n=== 비교 결과 ===")
    # pprint(comparison)

    # 방법 4: 여러 URL 일괄 처리
    # test_urls = [
    #     "https://arxiv.org/html/2505.24090v1",
    #     "https://arxiv.org/html/1905.11946v1"
    # ]
    # batch_results = batch_summarize_and_translate(test_urls, 100, 300, method="extractive", delay=10)
    # for i, result in enumerate(batch_results):
    #     print(f"\n=== URL {i+1} 결과 ===")
    #     pprint(result)
````
